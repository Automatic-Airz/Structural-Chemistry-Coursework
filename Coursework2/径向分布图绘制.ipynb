```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ D(r)=r^2R^2(r) $$

```python

delta = 0.01
r = np.arange(0, 1.5, delta)

a0 = 0.05292
# 单位：nm
sigma = r / a0

D2p =  r**2 * sigma**2 * np.exp(-sigma) / a0**3 /24 

D3p= r**2*(8/27/np.sqrt(6)*(1-sigma/6)*sigma*np.exp(-sigma/3))**2/a0**3

D3d = r**2 * (4/81/np.sqrt(30)*sigma**2 * np.exp(-sigma/3))**2 /a0**3

D4p = r**2 * (np.sqrt(5/3)/16*(1-sigma/4+sigma**2/80)* sigma * np.exp(-sigma/4))**2 /a0**3

plt.plot(r,D2p,label = "2p")
plt.plot(r,D3p,label = "3p")
plt.plot(r,D3d,label="3d")
plt.plot(r,D4p,label = "4p")

plt.legend()
plt.xlabel("r (nm)")
plt.ylabel("D (nm^-1)")
plt.title("Radial Distribution")
print(np.sum(D2p)*0.01)

plt.savefig("Radial Distribution.jpg",dpi=300)
plt.show()
```
