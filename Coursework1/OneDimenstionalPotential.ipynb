```python
#偶宇称波函数

#导入numpy库
import numpy as np
#导入绘图工具库
import matplotlib.pyplot as plt


x=np.linspace(0.001,4*np.pi,200,endpoint=True)
#对x进行tan计算
tan= np.tan(x)

#划分曲线，分别设置曲线样式
ym1 = np.ma.masked_where(tan > 0, tan)
ym2 = np.ma.masked_where(tan < 0, tan) 
#绘制曲线
lines = plt.plot(x,tan,x,ym1)
lines2 = plt.plot(x,ym2,label='tan')
#用红色虚线绘制正切函数竖直渐进线
plt.setp(lines[0], color='red',linestyle='dashed',linewidth=1)
#用蓝色线绘制曲线
plt.setp(lines[1], color='blue',linewidth=1)
plt.setp(lines2, color='blue',linewidth=1)
#设定x轴范围
plt.xlim(0,4*np.pi*1.1)
#设定y轴范围
plt.ylim(-9,9)
#设置X轴刻度值
plt.xticks(
    [0,np.pi/2,np.pi,3*np.pi/2,2*np.pi,5*np.pi/2,3*np.pi,7*np.pi/2,4*np.pi],
    ['0',r'$\pi/2$',r'$\pi$',r'$3\pi/2$',r'$2\pi$',r'$5\pi/2$',r'$3\pi$',r'$7\pi/2$',r'$4\pi$']
        )

#设置Y轴刻度值
plt.yticks(list(range(-9,10,1)))
#通过plt.gca()获取坐标轴对象 然后设置属性
ax=plt.gca()
#隐藏top和right周
ax.spines['top'].set_color("none")
ax.spines['right'].set_color("none")
#把左下设置为0点
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))


z0=4
z02=8
z03=12
x2 = np.linspace(0.5,4,100,endpoint=True)
y1=np.sqrt(z0*z0/(x2*x2)-1)
function1 = plt.plot(x2,y1,label='y=(ε0^2/ε^2-1)^0.5, ε0=4',color='orange')

x2 = np.linspace(0.5,8,100,endpoint=True)
y2=np.sqrt(z02*z02/(x2*x2)-1)
function2 = plt.plot(x2,y2,label='y=(ε0^2/ε^2-1)^0.5, ε0=8',color='xkcd:turquoise')

x2 = np.linspace(0.5,12,100,endpoint=True)
y3=np.sqrt(z03*z03/(x2*x2)-1)
function3 = plt.plot(x2,y2,label='y=(ε0^2/ε^2-1)^0.5, ε0=12',color='xkcd:aqua')


ax.legend()



plt.savefig("Solution1.jpg",dpi = 300)
```
