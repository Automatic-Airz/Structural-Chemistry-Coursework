```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
from matplotlib.path import Path
from matplotlib.patches import PathPatch


# Fixing random state for reproducibility
np.random.seed(19680801)

#ref:https://matplotlib.org/stable/gallery/images_contours_and_fields/image_demo.html#sphx-glr-gallery-images-contours-and-fields-image-demo-py
```

```python
def PsiPlot(nx,ny):
    delta = 0.025
    x = y = np.arange(-3.0, 3.0, delta)
    X, Y = np.meshgrid(x, y)

    a = 6
    psi = 2 / a * np.sin(nx * np.pi * (X-a/2)/a) * np.sin(ny * np.pi * (Y-a/2)/a)

    fig, ax = plt.subplots()
    im = ax.imshow(psi, interpolation='bilinear', cmap=cm.Spectral,
                origin='lower', extent=[-3, 3, -3, 3],
                vmax=abs(psi).max(), vmin=-abs(psi).max())
    print("量子数nx="+str(nx))
    print("量子数ny="+str(ny))
    print("\n")

    plt.show()
```

```python
for nx in range(1,4):
    for ny in range(1,4):
        PsiPlot(nx, ny)
```

```python
fig, axs = plt.subplots(2, 3, sharex=True, sharey=True)
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)

# Todo: 加个label 

axs[0,0].set_ylabel('y')
axs[1,0].set_ylabel('y')

axs[1,0].set_xlabel('x')
axs[1,1].set_xlabel('x')
axs[1,2].set_xlabel('x')

a = 6
for nx in range(1,4):
    for ny in range(1,5-nx):
        psi = 2 / a * np.sin(nx * np.pi * (X+a/2)/a) * np.sin(ny * np.pi * (Y+a/2)/a)
        if nx == 3:
            im = axs[1,2].imshow(psi, interpolation='bilinear', cmap=cm.Spectral,
            origin='lower', extent=[-3, 3, -3, 3],
            vmax=abs(psi).max(), vmin=-abs(psi).max())
            axs[1,2].set_title("nx="+str(nx)+",ny="+str(ny))
        else:
            im = axs[nx-1,ny-1].imshow(psi, interpolation='bilinear', cmap=cm.Spectral,
            origin='lower', extent=[-3, 3, -3, 3],
            vmax=abs(psi).max(), vmin=-abs(psi).max())
            axs[nx-1,ny-1].set_title("nx="+str(nx)+",ny="+str(ny))

cNorm = plt.Normalize(vmax=abs(psi).max(), vmin=-abs(psi).max())

# 加个 样条图例
fig.colorbar(plt.cm.ScalarMappable(cmap=cm.Spectral,norm=cNorm),ax=axs[:2],label = "Colorbar")

plt.savefig("WaveFunction.jpg",dpi = 300)
plt.show()

```
